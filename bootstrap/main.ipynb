```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
import random
```

```python
# Plotting Normal Distribution
mu = 2
variance = 0.25
sigma = np.sqrt(variance)
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma))
plt.show()
```

```python
# Simple Bootstrap Variance Estimation 
n = 1000
B = 100

Tn = []
for i in range(B):
    xhut = np.random.normal(mu, sigma, n)+np.random.uniform()
    t = np.mean(xhut)
    Tn.append(t)

var_boot = np.var(Tn)
print(var_boot)
```
